## #1:  Super-resolution imaging 

Obtaining several snapshots of the same scene, from microscopes to telescopes, is useful for the postprocessing increase of signal to noise: by summing up imaging data we can effectively beat down the noise. Interestingly, if we image the same scene from different vistas we can also improve the clarity of the combined image. Being able to discern features in a scene from this combination effort is sometimes called super-resolution imaging.

Here, we'll combine about 4 seconds of a shaky video to reveal the statement on a license plate that is not discernable in any one frame.

**Problem 1** Read in each image into a `numpy` array. Resize each frame to be 3 times larger in each axis (ie. 9 times larger images). Using `scipy.signal.fftconvolve` find the offsets of each frame with respect to the first frame. Report those offsets to 2 decimal places. 

   - Hint1: you'll need to figure out how to resize a numpy array
   - Hint2: you'll want to reverse the second image when doing the convolution: `scipy.signal.fftconvolve(im1, im2[::-1, ::-1])`
   - Hint3: you'll need to figure out how to identify the peak of the fft convolution to find the offsets between images

````python
import numpy as np
import scipy.ndimage
import glob

all_img = np.zeros((190*3,288*3,100))
imgs = glob.glob("license_plate_video/*")
img0 = scipy.ndimage.imread("license_plate_video/im2-0.png")[49:239,:,0]
img0_repeat = scipy.misc.imresize(img0,(img0.shape[0]*3,img0.shape[1]*3),interp='bilinear')
img0_repeat = img0_repeat - np.mean(img0_repeat)
offsets = np.zeros((len(imgs),2))
for i in range(len(imgs)):
    img = scipy.ndimage.imread(imgs[i])[49:239,:,0]
    img_repeat = scipy.misc.imresize(img,(img.shape[0]*3,img.shape[1]*3),interp='bilinear')
    img_repeat = img_repeat - np.mean(img_repeat)
    all_img[:,:,i] = img_repeat
    fft = scipy.signal.fftconvolve(img0_repeat,img_repeat[::-1,::-1])
    offset = np.array(np.unravel_index(fft.argmax(),fft.shape))-np.array(fft.shape)//2
    offsets[i] = offset
    print("image %i: offset %.2f, %.2f (x,y)" % (i, offset[0]/3, offset[1]/3))
````

**Problem 2** Shift each image to register the frames to the original (expanded in size) frame. You should, in general, be shifting by subpixel offsets. You might want to look at `scipy.ndimage.interpolation.shift`

````python
stacked_img = np.zeros(all_img[:,:,0].shape)
for i in range(len(imgs)):
    stacked_img += scipy.ndimage.interpolation.shift(all_img[:,:,i],offsets[i])
````

**Problem 3** Combine all the registered images to form a super-resolution image. What does the license plate read?

# #2: An elementary introduction to spectral audio compression

In this problem, we'll explore the very basics of audio compression in the spectral domain using numpy and scipy. We'll do a bit of visualization with matplotlib, but since that is covered later in the course, we'll provide those functions for you.

Audio compression is a large and complex topic, and the design of a format for compressed audio such as the popular [MP3](http://en.wikipedia.org/wiki/MP3) is too complex to cover in detail here. However, we will introduce the basic tools that most such compression formats use, namely:

1. Converting the input signal to the frequency domain by taking a Fast Fourier Transform (FFT).

2. Dropping information in the frequency domain, resulting in a smaller amount of data.

3. Reconstructing back the signal in the time domain from this smaller representation of the signal.

Steps 1 and 2 above are the 'encoding' part of signal compression, and step 3 is the 'decoding' part. For this reason, the tools that perform these steps are typically referred to as signal 'codecs', short for encoders/decoders.

Note that here we say 'signal': while MP3 is an audio format, the same ideas apply to the compression of digital images with formats such as JPEG and video.  Virtually all multimedia technologies we use today, from audio players to cell phones, digital cameras and YouTubeVideo, are based on sophisticated extensions and applications of these simple ideas.

Let's first load the plotting tools and importing some tools we'll need later:

````python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt

# we'll need some path manipulations later on
import os
````

We define a simple utility function to listen to audio files right in the browser:

````python
def Audio(fname):
    """Provide a player widget for an audio file.
    
    Parameters
    ==========
    fname : string
      Filename to be played.
      
    Warning
    =======
    
    Browsers cache audio very aggressively. If you change an
    audio file on disk and are trying to listen to the  new version, you 
    may want to 
    """
    from IPython.display import HTML, display
    
    # Find out file extension and deduce MIME type for audio format
    ext = os.path.splitext(fname)[1].replace('.', '').lower()
    mimetype = 'audio/' + ('mpeg' if ext == 'mp3' else ext)
    
    tpl = """<p>{fname}:</p>
<audio controls>
    <source src="files/{fname}" type="{mimetype}">

Your browser does not support the Audio element; you can play 
<a href="files/{fname}">this file</a> manually.
</audio>
"""
    display(HTML(tpl.format(**locals())))
````

We also define a convenience wrapper around `plt.specgram`, [matplotlib's spectrogram function](http://matplotlib.org/api/mlab_api.html#matplotlib.mlab.specgram), with a colorbar and control over the color limits displayed. This will make it easier to compare across different signals with the same colors for all inputs.

````python
def specgram_cbar(x, title=None, clim=(0, 80) ):
    """Plot spectrogram with a colorbar and range normalization.
    
    Call matplotlib's specgram function, with a custom figure size, 
    automatic colobar, title and custom color limits to ease 
    comparison across multiple figures.
    
    Parameters
    ==========
    x : array
      One-dimensional array whose spectrogram should be plotted.
      
    title : string
      Optional title for the figure.
      
    clim : 2-tuple
      Range for the color limits plotted in the spectrogram.
    """
    f = plt.figure(figsize=(10,3))
    plt.specgram(x)
    plt.colorbar()
    plt.clim(*clim)
    if title is not None:
        plt.title(title)
    plt.show()
````

**Problem 2**: Write a function to compress a 1-d signal by dropping a fraction of its spectrum. 

You can drop the smallest components by setting their values to zero.

*Hints*: 

- look at the `np.fft` module, keeping in mind that your input signal is real.
- look at the `argsort` method of numpy arrays.

````python
def compress_signal(x, fraction):
    """Compress an input signal by dropping a fraction of its spectrum.
    
    Parameters
    ==========
    x : array
      1-d real array to be compressed
      
    fraction : float
      A number in the [0,1] range indicating which fraction of the spectrum
      of x should be zeroed out (1 means zero out the entire signal).
      
    Returns
    =======
    x_approx : array
      1-d real array reconstructed after having compressed the input.
    """
    signal = x - np.mean(x)
    fft_signal = np.fft.rfft(signal)
    fft_mag = np.abs(fft_signal)
    inds = fft_mag.argsort()
    fft_signal[inds[:int(fft_signal.shape[0]*fraction)]] = 0.0
    return np.fft.irfft(fft_signal) + np.mean(x)
````

As a quick visual check (not that this is *not* a formal test of correctness), experiment with a simple random signal by changing the compression ratio and plotting both the signal and the compressed version:

````python
x = np.random.rand(128)
````

````python
fraction = 0.2  # play changing this in the 0-1 range

xa = compress_signal(x, fraction)

plt.figure(figsize=(12,3))
plt.plot(x, alpha=0.5, lw=2, label='original')
plt.plot(xa, lw=2, label='compressed {0:.0%}'.format(fraction))
plt.legend();
````

**Problem 3**: Write a function that will compress an audio file by a dropping a fraction of its spectrum, writing the output to a new file.

If the input file is named `a.wav` and the compression fraction is 0.9, the output file should be named `a_comp_0.9.wav`.

*Hints:* 

- look at the `scipy.io` module for routines dealing with files in `wav` format.

- you may need to use the `astype` method of numpy arrays to get the correct data type for `wav` files.

````python
import scipy.io.wavfile

def compress_wav(fname, fraction):
    """Compress an audio signal stored in an input wav file.
    
    The compressed signal is returned as a numpy array and automatically written 
    to disk to a new wav file.
    
    Parameters
    ==========
    fname : string
      Name of the input wav file
      
    fraction : float
      Fraction of input data to keep.
      
    Returns
    =======
    rate : int
      Bit rate of the input signal.

    x : array
      Raw data of the original input signal.
      
    x_approx : array
      Raw data of the compressed signal.
      
    new_fname : string
      Auto-generated filename of the compressed signal.
    """
    
    rate, data = scipy.io.wavfile.read(fname)
    comp_sig = compress_signal(data,fraction).astype("int16")
    new_fname = "".join(fname.split(".wav")[:-1])+"_comp_%.2f.wav" % fraction
    scipy.io.wavfile.write(new_fname,rate,comp_sig)
    return rate, data, comp_sig, new_fname
````

**Problem 4**: Study the effect of compressing the input file at different ratios: 0.1, 0.5, 0.75, 0.9, 0.95, 0.99.

Using the `OrderedDict` class from the [Python collections module](http://docs.python.org/2/library/collections.html#collections.OrderedDict), store the uncompressed signal as well as the compressed array and filename for each compression ratio.

You will create an `OrderedDict` called `voices`, with:

- keys: compression ratios
- values: pairs of (x, filename) where x is the compressed audio and filename is the name of the compressed file.

````python
import collections

comp_ratios = [0.1,0.5,0.75,0.9,0.95,0.99]
file = "voice.wav"
voices = collections.OrderedDict()
for comp in comp_ratios:
    rate, x, x_approx, new_fname = compress_wav(file,comp)
    voices[comp] = (x_approx, new_fname)
````

I had some trouble with the Lomb-scargle periodogram...the location of the peak frequency seems to depend a lot on the frequencies tested (see the two examples with 100 and 1000 evenly-spaced frequencies between 1 and 100).  So I went with a plain-old Fourier transform, which should be fine for this data as it is regularly spaced (a Lomb-Scargle periodogram is apparently designed for irregularly-sampled data).  The peak of the FFT is at 0.238 weeks^-1 corresponding to a period of 29.4 days.  Interestingly, this is quite close to the period of the phases of the Moon (29.53 days) which is not the same as the Moon's orbital period because the Earth moves on its orbit during the Moon's orbit.  As one would expect we can only recover the period of the phases, not the sidereal period ofthe Moon of 27.3 days.

````python
import numpy as np
import pandas as pd

alldata = pd.read_csv("merged_data.csv")
dates = pd.to_datetime(alldata.Week.values) - pd.datetime(2000,1,1)
x = dates.days.astype("float64")
y = alldata["ice cream"].values.astype("float64")

fft = np.fft.rfft(y-np.mean(y))
freq = np.fft.rfftfreq(y.size)
plt.plot(freq,np.abs(fft))
print("max frequency %f weeks^-1" % freq[np.argmax(np.abs(fft))])
````

I used "ice cream" as I expected it to be the best tracer of the year (it should peak around the same time each summer).  This tells me that a year is 52.2 weeks = 365.4 days, very close to the actual value of 365.25 days.

# #4: Reproducing some insights about the Election

Nate ("not a genius, just a Bayesian") Silver writes often about polls and their utility of predicting elections. One of the things he emphasized during the 2016 campaign is that even "large" polls of people with a consistent lead for one candidate will show wild swings in any given window in time.

**Problem 1**: Using Pandas and `numpy`, try to reproduce this plot from a Nate Silve Tweet qualitatively using the same assumptions.

<img src="https://www.evernote.com/l/AUU-KkH3Z-lBdKokzCzVIUe53YZLQ9e_AuMB/image.png">

https://twitter.com/NateSilver538/status/769565612955824128

````python
import numpy as np

def moving_avg(x,n):
    cumsum = np.cumsum(x)
    cumsum[n:] = cumsum[n:] - cumsum[:-n]
    return cumsum[n-1:]/n

n = 200.

sigma_diff = 100/np.sqrt(n) 

margin = 6
# Assumed Clinton margin

npolls = 150
np.random.seed()
polls = np.random.normal(loc=6,scale=sigma_diff,size=npolls)

# Plot a moving average of the polls rather than just the 200 person polls
plt.ylim([-3,15])
plt.yticks([-3,0,3,6,9,12])
plt.grid()
plt.plot(moving_avg(polls,5))
print("std %f" % np.std(moving_avg(polls,5)))
````

**Problem 2**: Clearly, even with a 6% point lead, there's a chance that this sort of poll would show the other person in the lead. How much would ahead (in percent) would a candidate need to be to have a tracking poll never show the other candidate to be ahead over the course of a year (in your simulation)?

````python
def tracking_poll(margin,n,npolls):
    sigma_diff = 100/np.sqrt(n) 
    polls = np.random.normal(loc=margin,scale=sigma_diff,size=npolls)
    return moving_avg(polls,5)

margins = np.linspace(4,15,12)
for m in margins:
    print("margin %.1f, minimum lead: %.2f" % (m,np.min(tracking_poll(m,200,365))))
````

Obviously there will be random variation from realization to realization, and in any given year, either candidate could be ahead in one poll regardless of the margin!  Nevertheless we can qualitatively see that once the lead exceeds 8 or 9 points, the likelihood of a 1 tracking poll in the year showing the other candidate ahead becomes very small.

**Problem 3**: With a 3 and 6% lead, how many people would need to be polled in 1 day to have the rolling 5-day poll result always show the leader ahead (over a year)?

````python
ns = np.logspace(2,4,30)
for n in ns:
    print("n %i, minimum lead: %.2f" % (n,np.min(tracking_poll(3,n,365))))
````
